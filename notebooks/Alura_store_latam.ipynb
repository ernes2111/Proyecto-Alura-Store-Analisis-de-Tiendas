### Importación de datos



```python
# Importacion de librerias y carga de datos
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head(10)
```

# 3. Calificación promedio de la tienda


```python
# Creamos un diccionacionario para almacenar los DataFrames de cada tienda
tiendas = {
    'Tienda 1': tienda,
    'Tienda 2': tienda2,
    'Tienda 3': tienda3,
    'Tienda 4': tienda4
}
# Creamos un diccionario para almacenar los promedios de calificaciones
promedios_calificaciones = {}

for nombre, df in tiendas.items():
    promedios_calificaciones[nombre] = df['Calificación'].mean()

print(promedios_calificaciones)
```

```python
# Tienda 3 tiene el mayor promedio de calificaciones
mejor_tienda = max(promedios_calificaciones, key=promedios_calificaciones.get)
print(f'La tienda con mejor promedio de calificaciones es {mejor_tienda} con un promedio de {promedios_calificaciones[mejor_tienda]:.2f}')
```

```python
import matplotlib.pyplot as plt

tiendas_nombres = list(promedios_calificaciones.keys())
valores = list(promedios_calificaciones.values())

plt.figure(figsize=(8,5))
plt.barh(tiendas_nombres, valores, color='skyblue', edgecolor='black')

for i, v in enumerate(valores):
    plt.text(v + 0.02, i, f"{v:.2f}", va='center', fontsize=10)

plt.title("Calificación promedio por tienda", fontsize=14)
plt.xlabel("Calificación promedio (0 a 5)")
plt.xlim(0, 5)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()

```

## Comparativa producto mas vendido por tienda

```python
#Creao lista con el producto más vendido por tienda
productos_top = []

for nombre, df in tiendas.items():
    # Agrupacion por producto y contamos las ventas
    ventas = df.groupby('Producto')['Precio'].count().sort_values(ascending=False)
    producto_top = ventas.head(1)  # top 1 producto más vendido
    
    productos_top.append({
        'Tienda': nombre,
        'Producto': producto_top.index[0],
        'Ventas': producto_top.values[0]
    })

#Convertimos a DataFrame
df_top = pd.DataFrame(productos_top)
print("Top productos más vendidos por tienda:")
print(df_top)

#Crear el gráfico Radar Chart
labels = df_top['Tienda']
values = df_top['Ventas']

#Cerrar el círculo del radar
angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()
values = values.tolist()
values += values[:1]
angles += angles[:1]

#Crear figura polar
fig, ax = plt.subplots(figsize=(7,7), subplot_kw=dict(polar=True))

#Dibujar línea y área
ax.plot(angles, values, linewidth=2, linestyle='solid', color='dodgerblue')
ax.fill(angles, values, color='skyblue', alpha=0.4)

#Etiquetas y detalles
ax.set_thetagrids(np.degrees(angles[:-1]), labels)
ax.set_title('Comparativa de productos más vendidos por tienda (Radar Chart)', fontsize=12, pad=20)
ax.grid(True, alpha=0.3)

#Mostrar los valores en el gráfico
for i, v in enumerate(df_top['Ventas']):
    ax.text(angles[i], v + (v*0.05), f"{v}", ha='center', va='center', fontsize=10, color='navy')

plt.show()
```

```python
# Re usamos el diccionario "tiendas" para analizar los productos más y menos vendidos por tienda, creado en el punto anterior
for nombre, df in tiendas.items():
    # Agrupar por producto y contar ventas
    ventas = df.groupby('Producto')['Precio'].count().sort_values(ascending=False)
    
    # Seleccionar top 3 y bottom 3
    top3 = ventas.head(3)
    bottom3 = ventas.tail(3)
    
    # Convertimos a DataFrame para mostrar mas bonito
    df_top = pd.DataFrame(top3).reset_index()
    df_top.columns = ['Producto', 'Cantidad de Ventas']
    
    df_bottom = pd.DataFrame(bottom3).reset_index()
    df_bottom.columns = ['Producto', 'Cantidad de Ventas']
    
    # Imprimir resultados
    print(f'\n===== {nombre} =====')
    print('\nTop 3 Productos más vendidos:')
    print(df_top.to_string(index=False))  # Sin índice para mejor presentación
    
    print('\n3 Productos con menos ventas:')
    print(df_bottom.to_string(index=False))

```

```python
for nombre, df in tiendas.items():
    ventas = df.groupby('Producto')['Precio'].count().sort_values(ascending=False)
    top3 = ventas.head(3)
    bottom3 = ventas.tail(3)
    
    # Unimos ambos
    df_plot = pd.concat([
        pd.DataFrame({'Producto': top3.index, 'Ventas': top3.values, 'Tipo': 'Top 3'}),
        pd.DataFrame({'Producto': bottom3.index, 'Ventas': bottom3.values, 'Tipo': 'Bottom 3'})
    ])
    
    fig, ax = plt.subplots(figsize=(7,5))
    for tipo, grupo in df_plot.groupby('Tipo'):
        ax.bar(grupo['Producto'], grupo['Ventas'], label=tipo, alpha=0.7)
    
    ax.set_title(f'{nombre} — Comparativa Top vs Bottom 3')
    ax.set_ylabel('Cantidad de Ventas')
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.xticks(rotation=45)
    plt.show()
```

# 5. Envío promedio por tienda

```python
# Calculamos el costo promedio de envio por tienda, usando el diccionario "tiendas"
promedios_envio = {}

for nombre, df in tiendas.items():
    promedios_envio[nombre] = df['Costo de envío'].mean()

#Imprimimos el costo promedio de envio por tienda con 2 decimales
for nombre, valor in promedios_envio.items():
    print(f'Costo promedio de envio {nombre}: ${valor:.2f}')
```

```python
promedios_envio = {
    'Tienda 1': 26018.61,
    'Tienda 2': 25216.24,
    'Tienda 3': 24805.68,
    'Tienda 4': 23459.46
}

# Ordenamos de mayor a menor para que se vea mas claro
promedios_envio = dict(sorted(promedios_envio.items(), key=lambda x: x[1], reverse=True))

tiendas_nombres = list(promedios_envio.keys())
valores = list(promedios_envio.values())

# Escala de colores según el valor
colors = plt.cm.coolwarm([(v - min(valores)) / (max(valores) - min(valores)) for v in valores])

plt.figure(figsize=(9,5))
bars = plt.barh(tiendas_nombres, valores, color=colors, edgecolor='black')

#para que quede en decimales
for i, v in enumerate(valores):
    plt.text(v + 200, i, f"${v:,.2f}", va='center', fontsize=10)

plt.title("Costo promedio de envío por tienda", fontsize=14, pad=15)
plt.xlabel("Costo promedio ($)")
plt.gca().invert_yaxis()  # Para que la tienda más cara quede arriba
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
